```python
#import library 
import pandas as pds
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##read data

data1=pds.read_csv('Dataset/Data_of_Attack_Back_BufferOverflow.csv')
data2=pds.read_csv('Dataset/Data_of_Attack_Back_FTPWrite.csv')
data3=pds.read_csv('Dataset/Data_of_Attack_Back_GuessPassword.csv')
data4=pds.read_csv('Dataset/Data_of_Attack_Back_Neptune.csv')
data5=pds.read_csv('Dataset/Data_of_Attack_Back_NMap.csv')
data6=pds.read_csv('Dataset/Data_of_Attack_Back_Normal.csv')
data7=pds.read_csv('Dataset/Data_of_Attack_Back_PortSweep.csv')
data8=pds.read_csv('Dataset/Data_of_Attack_Back_RootKit.csv')
data9=pds.read_csv('Dataset/Data_of_Attack_Back_Satan.csv')
data10=pds.read_csv('Dataset/Data_of_Attack_Back_Smurf.csv')
data11=pds.read_csv('Dataset/Data_of_Attack_Back.csv')
```

```python
# check column and row 
print('data1', data1.shape)
print('data2',data2.shape)
print('data3',data3.shape)
print('data4',data4.shape)
print('data5',data5.shape)
print('data6',data6.shape)
print('data7',data7.shape)
print('data8',data8.shape)
print('data9',data9.shape)
print('data10',data10.shape)
print('data11',data11.shape)

#result = data. (row, column)
```

```python
##add header column to data3 file which name is 'Data_of_Attack_Back_FTPWrite.csv'
data_header = data1.columns
data2.columns = data_header  #take header from data1 file to replace the header data2 file
```

```python
#add new column name which is the 42th column name = attack

data1['attack']=['BufferOverflow']*data1.shape[0]
data2['attack']=['FTPWrite']*data2.shape[0]
data3['attack']=['GuessPassword']*data3.shape[0]
data4['attack']=['Neptune']*data4.shape[0]
data5['attack']=['NMap']*data5.shape[0]
data6['attack']=['Normal']*data6.shape[0]
data7['attack']=['PortSweep']*data7.shape[0]
data8['attack']=['RootKit']*data8.shape[0]
data9['attack']=['Satan']*data9.shape[0]
data10['attack']=['Smurf']*data10.shape[0]
data11['attack']=['Back']*data11.shape[0]
```

```python
#ដាក់​ data ទាំង១១ file ចូលតែ 1 file
all_data=pds.concat([data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11])
all_data.reset_index()

#extract to new file which name is 'Data_of_Attack_all_data.csv'
all_data.to_csv('Dataset/Data_of_Attack_all_data.csv', index=False)

all_data=pds.read_csv('Dataset/Data_of_Attack_all_data.csv')
all_data.columns[0]
```

Data Scaling and Transformation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

# Create a copy of the 'attack' column and convert it to numerical representation
le = LabelEncoder()
all_data['attack_num'] = le.fit_transform(all_data['attack'])

# Select the numerical 'attack_num' column for scaling
data_scaling = all_data[['attack_num']].astype(float)

# Apply StandardScaler for data scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_scaling)

# Create a new DataFrame with scaled data
scaled_data_scaling = pds.DataFrame(scaled_data, columns=data_scaling.columns)

# Display the scaled DataFrame
print("Scaled Data:")
print(scaled_data_scaling)
```
